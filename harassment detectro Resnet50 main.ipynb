</div>

```python
# Function to parse XML annotation file and extract labels and bounding boxes
def parse_xml(file_path):
    """
    Parse XML annotation file and extract labels and bounding boxes.
    
    Args:
        file_path (str): Path to XML annotation file
    
    Returns:
        tuple: Lists of labels and bounding boxes
    """
    try:
        # Parse the XML tree and get the root element
        tree = ET.parse(file_path)
        root = tree.getroot()

        labels = []  # List to store labels
        bboxes = []  # List to store bounding boxes

        # Iterate over each 'object' in the XML
        for obj in root.findall('object'):
            label = obj.find('name').text.lower()  # Get the object label
            
            # Map label to numeric value using LABEL_MAP
            if label in LABEL_MAP:
                label_index = LABEL_MAP[label]
                
                # Extract bounding box coordinates
                bbox = obj.find('bndbox')
                xmin = int(float(bbox.find('xmin').text))
                ymin = int(float(bbox.find('ymin').text))
                xmax = int(float(bbox.find('xmax').text))
                ymax = int(float(bbox.find('ymax').text))

                # Append label and bounding box to respective lists
                labels.append(label_index)
                bboxes.append((xmin, ymin, xmax, ymax))

        return labels, bboxes
    except Exception as e:
        print(f"Error parsing XML {file_path}: {e}")
        return [], []  # Return empty lists in case of error

# Function to load image, crop according to bounding boxes, and preprocess
def load_and_preprocess_image(image_path, bboxes, target_size=(224, 224)):
    """
    Load image, crop bounding boxes, and preprocess.
    
    Args:
        image_path (str): Path to image file
        bboxes (list): List of bounding boxes
        target_size (tuple): Resize dimensions
    
    Returns:
        list: Preprocessed image crops
    """
    try:
        # Read the image from file
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"Cannot read image: {image_path}")

        # Convert BGR image to RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        h, w = image.shape[:2]  # Get image height and width

        processed_images = []  # List to store processed image crops

        # Iterate over each bounding box to crop and process the image
        for bbox in bboxes:
            xmin, ymin, xmax, ymax = bbox
            
            # Validate bounding box to ensure it lies within image dimensions
            xmin = max(0, min(xmin, w))
            ymin = max(0, min(ymin, h))
            xmax = max(0, min(xmax, w))
            ymax = max(0, min(ymax, h))

            # Crop the image using bounding box coordinates
            crop = image[ymin:ymax, xmin:xmax]
            
            # Convert the cropped image to PIL format and resize it
            pil_image = Image.fromarray(crop)
            pil_image = pil_image.resize(target_size)
            
            # Append the processed crop to the list
            processed_images.append(pil_image)

        return processed_images
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return []  # Return an empty list in case of error
```

```python
# Custom Dataset class for harassment detection task
class HarassmentDataset(Dataset):
    def __init__(self, image_paths, labels, bboxes, transform=None):
        """
        Initialize the dataset with image paths, labels, bounding boxes, and optional transformations.
        
        Args:
            image_paths (list): List of paths to image files
            labels (list): List of corresponding labels for each image
            bboxes (list): List of bounding boxes for each image
            transform (callable, optional): A function/transform to apply to the images
        """
        self.image_paths = image_paths  # Store image paths
        self.labels = labels            # Store labels
        self.bboxes = bboxes            # Store bounding boxes
        self.transform = transform      # Store transformation function, if provided

    def __len__(self):
        """
        Returns the number of samples in the dataset.
        """
        return len(self.image_paths)

    def __getitem__(self, idx):
        """
        Fetches a single sample from the dataset.
        
        Args:
            idx (int): Index of the sample to fetch
        
        Returns:
            tuple: Processed image(s) and corresponding label
        """
        image_path = self.image_paths[idx]  # Get image path for the given index
        bboxes = self.bboxes[idx]           # Get bounding boxes for the image
        label = self.labels[idx]            # Get the label for the image

        # Load and preprocess the image and bounding boxes
        images = load_and_preprocess_image(image_path, bboxes)
        
        if not images:
            return None  # If no images are processed, return None

        # Apply the transformation (if any) to the images
        if self.transform:
            images = [self.transform(img) for img in images]

        # If there are multiple crops, take the first one
        if len(images) > 1:
            images = images[0]
        else:
            images = images[0]

        return images, label  # Return the image(s) and the corresponding label


# Updated custom collate function for handling variable batch sizes and image crops
def custom_collate(batch):
    """
    Custom collate function to handle variable batch sizes and inconsistent image crops.
    
    Args:
        batch (list): List of samples in a batch, each a tuple of (image, label)
    
    Returns:
        tuple: Batch of images and corresponding labels, as tensors
    """
    # Filter out None values from the batch (in case any sample returned None)
    batch = [item for item in batch if item is not None]
    
    if not batch:
        return None  # If the batch is empty after filtering, return None
    
    # Separate images and labels
    images, labels = zip(*batch)
    
    # Stack images into a single tensor (assuming they are all the same size)
    images = torch.stack(images)
    
    # Convert labels to a tensor
    labels = torch.tensor(labels)
    
    return images, labels  # Return the stacked images and corresponding labels
```

```python
# Function to load dataset from the specified directory
def load_dataset(dataset_dir):
    """
    Load dataset from directory containing images and annotations.
    
    Args:
        dataset_dir (str): Path to dataset directory which contains 'Annotations' and 'JPEGImages' subdirectories.
    
    Returns:
        tuple: Lists of image paths, labels, and bounding boxes
    """
    # Define paths to annotations and images directories
    annotations_dir = os.path.join(dataset_dir, "Annotations")
    images_dir = os.path.join(dataset_dir, "JPEGImages")

    # Initialize lists to store image paths, labels, and bounding boxes
    image_paths = []
    all_labels = []
    all_bboxes = []

    # Iterate through all XML files in the annotations directory
    for xml_file in os.listdir(annotations_dir):
        if xml_file.endswith('.xml'):  # Only process XML files
            xml_path = os.path.join(annotations_dir, xml_file)
            
            # Get the corresponding image name and path (assumes .png images)
            image_name = os.path.splitext(xml_file)[0] + '.png'
            image_path = os.path.join(images_dir, image_name)

            if not os.path.exists(image_path):  # Skip if image does not exist
                continue

            # Parse the XML file to get labels and bounding boxes
            labels, bboxes = parse_xml(xml_path)
            
            if labels:  # Only add to dataset if labels are valid
                image_paths.append(image_path)
                all_labels.append(labels[0])  # Only take the first label
                all_bboxes.append(bboxes)

    return image_paths, all_labels, all_bboxes  # Return image paths, labels, and bounding boxes

# Transformations for training data
transform_train = transforms.Compose([
    transforms.RandomResizedCrop(224),   # Randomly crop and resize images to 224x224
    transforms.RandomHorizontalFlip(),    # Randomly flip images horizontally
    transforms.ColorJitter(brightness=0.2, contrast=0.2),  # Randomly adjust brightness and contrast
    transforms.RandomRotation(10),        # Randomly rotate images by up to 10 degrees
    transforms.ToTensor(),                # Convert images to PyTorch tensors
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize images
])

# Transformations for testing data
transform_test = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),  # Resize images to a fixed size
    transforms.ToTensor(),                    # Convert images to PyTorch tensors
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # Normalize images
])
```

```python
from torchvision.models import resnet50
import torch.nn as nn

class HarassmentClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        """
        Initialize the HarassmentClassifier model.
        
        Args:
            num_classes (int): Number of output classes (default is 3)
        """
        super(HarassmentClassifier, self).__init__()
        
        # Load the pre-trained ResNet50 backbone
        self.backbone = resnet50(pretrained=True)
        
        # Get the number of input features to the final fully connected layer
        num_features = self.backbone.fc.in_features

        # Replace the fully connected layer with a custom head
        self.backbone.fc = nn.Sequential(
            # First layer: Linear layer with 2048 units, followed by ReLU, BatchNorm, and Dropout
            nn.Linear(num_features, 2048),
            nn.ReLU(),
            nn.BatchNorm1d(2048),
            nn.Dropout(0.5),
            
            # Second layer: Linear layer with 1024 units, followed by ReLU, BatchNorm, and Dropout
            nn.Linear(2048, 1024),
            nn.ReLU(),
            nn.BatchNorm1d(1024),
            nn.Dropout(0.4),
            
            # Third layer: Linear layer with 512 units, followed by ReLU, BatchNorm, and Dropout
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.3),
            
            # Final output layer with number of classes
            nn.Linear(512, num_classes)
        )
    
    def forward(self, x):
        """
        Forward pass through the model.
        
        Args:
            x (tensor): Input tensor
        
        Returns:
            tensor: Output predictions
        """
        return self.backbone(x)
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau

def train_model(model, train_loader, val_loader, criterion, optimizer, epochs):
    """
    Train the model and evaluate it using validation data, while applying an LR scheduler.
    
    Args:
        model (nn.Module): The model to be trained.
        train_loader (DataLoader): DataLoader for the training set.
        val_loader (DataLoader): DataLoader for the validation set.
        criterion (nn.Module): The loss function.
        optimizer (Optimizer): The optimizer for training.
        epochs (int): Number of epochs to train the model.
    
    Returns:
        dict: A dictionary containing training history including losses, accuracies, and learning rates.
    """
    # Check if a GPU is available, otherwise use CPU
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # Initialize learning rate scheduler
    scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5)

    # Dictionary to store the training history
    train_history = {
        'train_loss': [],
        'val_loss': [],
        'train_acc': [],
        'val_acc': [],
        'lr': []  # Track learning rate changes
    }

    # Training loop
    for epoch in range(epochs):
        model.train()  # Set the model to training mode
        train_loss = 0.0
        train_correct = 0
        train_total = 0

        # Iterate through training batches
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()  # Zero gradients for each batch
            outputs = model(images)  # Forward pass
            loss = criterion(outputs, labels)  # Compute loss
            loss.backward()  # Backward pass (gradient computation)
            optimizer.step()  # Optimizer step to update weights

            # Track loss and accuracy
            train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)  # Get the predicted class
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()

        # Validation phase
        model.eval()  # Set the model to evaluation mode
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():  # Disable gradient computation for validation
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        # Compute average losses and accuracies
        train_loss_avg = train_loss / len(train_loader)
        val_loss_avg = val_loss / len(val_loader)
        train_acc = train_correct / train_total
        val_acc = val_correct / val_total

        # Store the results for each epoch
        train_history['train_loss'].append(train_loss_avg)
        train_history['val_loss'].append(val_loss_avg)
        train_history['train_acc'].append(train_acc)
        train_history['val_acc'].append(val_acc)

        # Update the learning rate scheduler
        scheduler.step(val_loss_avg)
        current_lr = optimizer.param_groups[0]['lr']
        train_history['lr'].append(current_lr)

        # Print summary for the epoch
        print(f'Epoch {epoch+1}/{epochs}')
        print(f'Train Loss: {train_loss_avg:.4f}, Train Acc: {100*train_acc:.2f}%')
        print(f'Val Loss: {val_loss_avg:.4f}, Val Acc: {100*val_acc:.2f}%')
        print(f'Learning Rate: {current_lr:.6f}')

    # Return training history for later analysis
    return train_history
```
